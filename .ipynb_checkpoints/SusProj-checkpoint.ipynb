```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# The below commands make the font and image size bigger
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

# Chegg Example 

```python
def tester(t,y):
    return [y[1],(-y[1]**2)/y[0]-0.001/y[0]]

initderiv = np.linspace(0.04,0.05,20)

for el in initderiv:
    sol =solve_ivp(tester,[0,1000],[10,el],rtol=1e-12,atol = 1e-12)
    y = sol.y
    t = sol.t
    plt.plot(t,y[0])
plt.scatter(1000,5,s=100)
```
